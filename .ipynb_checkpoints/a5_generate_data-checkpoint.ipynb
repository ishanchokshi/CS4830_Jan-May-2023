```python
import pandas as pd
import numpy as np
import random

# Define the lists of data
product_name_list = ["Laptop", "Desktop Computer", "Mobile Phone", "Wrist Band", "Wrist Watch", "LAN Cable",
                     "HDMI Cable", "TV", "TV Stand", "Text Books", "External Hard Drive", "Pen Drive", "Online Course"]
order_card_type_list = ["Visa", "MasterCard", "Maestro"]
country_name_city_name_list = ["Sydney,Australia", "Florida,United States", "New York City,United States",
                               "Paris,France", "Colombo,Sri Lanka", "Dhaka,Bangladesh", "Islamabad,Pakistan",
                               "Beijing,China", "Rome,Italy", "Berlin,Germany", "Ottawa,Canada",
                               "London,United Kingdom", "Jerusalem,Israel", "Bangkok,Thailand",
                               "Chennai,India", "Bangalore,India", "Mumbai,India", "Pune,India",
                               "New Delhi,Inida", "Hyderabad,India", "Kolkata,India", "Singapore,Singapore"]
ecommerce_website_name_list = ["www.healthkart.com", "www.amazon.com", "www.flipkart.com", "www.snapdeal.com", "www.ebay.com"]

# Define the number of rows in the DataFrame
num_rows = 100

# Generate random data for each column
product_name = [random.choice(product_name_list) for _ in range(num_rows)]
order_card_type = [random.choice(order_card_type_list) for _ in range(num_rows)]
country_city = [random.choice(country_name_city_name_list) for _ in range(num_rows)]
order_country_name = [x.split(',')[1] for x in country_city]
order_city_name = [x.split(',')[0] for x in country_city]
order_amount = [round(np.random.uniform(10, 1000), 2) for _ in range(num_rows)]
order_datetime = [pd.Timestamp.now() - pd.to_timedelta(np.random.randint(1, 10000), unit='s') for _ in range(num_rows)]
order_datetime = [ts.round(freq='s') for ts in order_datetime]
order_ecommerce_website_name = [random.choice(ecommerce_website_name_list) for _ in range(num_rows)]
order_id = list(range(1, num_rows+1))

# Create the pandas DataFrame
df = pd.DataFrame({
    'order_id': order_id,
    'order_product_name': product_name,
    'order_card_type': order_card_type,
    'order_amount': order_amount,
    'order_datetime': order_datetime,
    'order_country_name': order_country_name,
    'order_city_name': order_city_name,
    'order_ecommerce_website_name': order_ecommerce_website_name
})
```

```python
df.to_csv('data.csv')
```
